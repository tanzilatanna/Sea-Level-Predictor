```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import csv
```

```python
#Task:- To import csv data using pandas
#Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.
```

```python
df = pd.read_csv('ESL.csv')
x=df['Year']
y=df['CSIRO Adjusted Sea Level']
plt.scatter(x,y)
```

```python
#Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. 
#Plot the line of best fit over the top of the scatter plot. 
```

```python
result =linregress(x,y)
x1 = pd.Series(i for i in range(1880,2051))
y1= result.slope* x1  + result.intercept
plt.plot(x1, y1,'r')
```

```python
#Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. 
#Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as 
#it has since the year 2000.
#The x label should be Year, the y label should be Sea Level (inches), and the title should be Rise in Sea Level.
```

```python
#creating second line of the best fit
newdf = df.loc[df['Year']>= 2000]
Xdf = newdf['Year']
Ydf= newdf['CSIRO Adjusted Sea Level']
res = linregress(Xdf,Ydf)
Xdf1 = pd.Series(i for i in range(2000,2050))
Ydf1= res.slope *Xdf1 + res.intercept
plt.plot(Xdf1, Ydf1)
plt.xlabel('Year')
plt.ylabel('Sea Level in Inches')
plt.title('Rise in Sea Level')
```
